##### Creando tablas de resumen y gráficos

```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
from matplotlib.pyplot import pie, axis, show
from matplotlib import cm
%matplotlib inline
```

¿Cuál nacionalidad, entre todos los astronautas, es la que más tiempo en promedio ha estado en el espacio?

```python
nueva = astro.groupby(['nationality']).agg(
{'total_hrs_sum': ['mean', 'max','min']}
).reset_index()
```

```python
nueva.columns = [''.join(col).strip() for col in nueva.columns.values]
```

```python
new_col = []
for col in nueva.columns.values:
    a = ''.join(col).strip()
    new_col.append(a)
nueva.columns = new_col
```

```python
nueva.sort_values('total_hrs_summean', inplace=True)
nueva
```

```python
nueva.sort_values('total_hrs_summean',ascending = False,  inplace = True)
```

```python
plt.figure(figsize=(8,8))
nueva['total_hrs_summean'].head(30).plot.bar()
plt.ylabel('Time in Space')
plt.xlabel('nationality')
```

```python
#Definiendo un tema
def tema_graf():
    font = "Lato"
    labelFont = "Garamond"
    sourceFont = "Calibri"
    
    return {
        "config" : {
             "title": {'font': font, 
                       'fontSize': 15,
                       'fontWeight': 300,
                       'offset':0,
                       'anchor':'left'},
             "axis": {
                  "labelFont": font,
                  "titleFont": font,
                  "domainColor": "gray"
             },
             "header": {
                  "labelFont": font,
                  "titleFont": font
             },
             "legend": {
                  "labelFont": labelFont,
                  "titleFont": labelFont,
                  "symbolType": 'circle',
                  "padding": 1
                  
             }
             ,
             'background': '#d7dff4'             
        }
    }

alt.themes.register('mitema', tema_graf)
alt.themes.enable('mitema')
```

```python
alt.Chart(nueva, height = 500, width = 800).mark_bar().encode(
    x = "total_hrs_summean",
    y = "nationality").properties(
    title={
      "text": ["Promedio de horas en el espacio por país"], 
      "subtitle": ["1961-2019",
                   "Fuente: NASA"],
      "color": "Black",
      "subtitleColor": "Black"}
    )
```
